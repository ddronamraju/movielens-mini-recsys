```python
from recommender.recommend import recommend
from recommender.data import load_processed_ratings

df = load_processed_ratings()
movies = df[["movie_id", "title"]].drop_duplicates().set_index("movie_id")

sample_user = int(df["user_id"].iloc[0])

print("Random:")
print(movies.loc[recommend(sample_user, k=5, model="random")]["title"].tolist())

print("\nPopularity:")
print(movies.loc[recommend(sample_user, k=5, model="popularity")]["title"].tolist())

print("\nSVD MF:")
print(movies.loc[recommend(sample_user, k=5, model="svd_mf")]["title"].tolist())
```
