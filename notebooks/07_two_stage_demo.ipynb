```python
from recommender.recommend import recommend
from recommender.data import load_processed_ratings

import pandas as pd

df = load_processed_ratings()
movies = df[["movie_id", "title"]].drop_duplicates().set_index("movie_id")

sample_user = int(df["user_id"].iloc[0])
sample_user

def print_recs(user_id, model, k=10):
    movie_ids = recommend(user_id, k=k, model=model)
    titles = movies.loc[movie_ids]["title"].tolist()
    print(f"Model: {model}")
    for i, t in enumerate(titles, 1):
        print(f"{i:2d}. {t}")
    print()

print_recs(sample_user, "popularity", k=10)
print_recs(sample_user, "svd_mf", k=10)
print_recs(sample_user, "two_stage", k=10)

for uid in df["user_id"].sample(3, random_state=42).tolist():
    print("="*40)
    print(f"User {uid}")
    print_recs(uid, "popularity", k=5)
    print_recs(uid, "svd_mf", k=5)
    print_recs(uid, "two_stage", k=5)
```
